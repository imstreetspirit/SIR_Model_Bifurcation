```python
import numpy as np
import matplotlib.pyplot as plt
from numpy import linalg as LA

#x, y = np.mgrid[2:-2:20j, 2:-2:20j]
x, y = np.meshgrid(np.linspace(-2, 2, 30), np.linspace(-2, 2, 30))
alpha = -0.01
dx = alpha*x - alpha*y
dy = -0.25*x 

w, v = LA.eig(np.array([[alpha, -alpha], [-0.25, 0]]))
 

plt.streamplot(x, y, dx, dy)
plt.plot(0,0, linestyle='--', marker='x', color='r')
plt.title(f"\u03B1: {alpha},\u03BB₁:{str(w[0])}, \u03BB₂: {str(w[1])} ")
plt.axis('square')
plt.axis([-2, 2, -2, 2])
plt.savefig("new_" + str(alpha) + ".png")
plt.show()
```
